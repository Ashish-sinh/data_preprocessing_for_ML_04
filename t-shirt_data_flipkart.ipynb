```python
import pandas as pd 
import requests 
from  bs4 import BeautifulSoup 
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"} 
url = 'https://www.flipkart.com/clothing-and-accessories/topwear/tshirt/men-tshirt/pr?sid=clo,ash,ank,edy&otracker=categorytree&otracker=nmenu_sub_Men_0_T-Shirts'

web_page = requests.get(url,headers).content 

soup = BeautifulSoup(web_page , 'lxml')
```

```python
prices = [] 
for i in soup.find_all('div',{'class':'_30jeq3'}): 
    prices.append(i.text)
```

```python
style = [] 
for i in soup.find_all('div',{'class':'_2WkVRV'}):
    if i == None : 
        style.append('0') 
    else : 
        style.append(i.text) 
```

```python
tshirt_names = [] 
for i in soup.find_all('a', class_= "IRpwTa"): 
    tshirt_names.append(i.text)
```

```python
discount = [] 
for i in soup.find_all('div',class_ = "_3Ay6Sb"): 
    discount.append(float(i.text[:2]))
```

```python
original_price = [] 
for i in soup.find_all('div',class_ = "_3I9_wc"):
    original_price.append(i.text)
```

```python
extra_feature =[] 
for i in soup.find_all('div',class_ = "_2Tpdn3"): 
    extra_feature.append(i.text)
```

```python
df = pd.DataFrame() 

df['tshirt_names'] = tshirt_names 
df['style']  = style
df['prices'] = prices [ :40]
df['discount'] =discount [:40]
df['original_price'] = original_price [:40]
df['extra_feature'] = extra_feature [:40]
```

```python
df['prices'] =df['prices'].str.replace('₹','')
df['prices'] =df['prices'].str.replace(',','')
df['prices'] =pd.to_numeric(df['prices'])

df['discount'] = pd.to_numeric(df['discount'])

df['original_price'] =df['original_price'].str.replace('₹','')
df['original_price'] =df['original_price'].str.replace(',','')
df['original_price'] =pd.to_numeric(df['original_price'])
```
